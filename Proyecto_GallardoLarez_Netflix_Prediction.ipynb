# Lectura de dataset

```python
df = pd.read_csv('/content/netflix_titles.csv')
```

# Agregar id como indice

```python
df.set_index('show_id',inplace=True)
df.head()
```

# Preprocesamiento de datos

Esto se hace debido a que en el dataset el campo está en blanco en la columna de director, para que no haya problemas al realizar las predicciones se llena los datos nulos con NaN especificando que no se agregó director.

```python
df.fillna('NaN')
```

```python
df.head()
```

Se utiliza label encoder para las columnas que se utilizarán para aplicar el modelo y realizar la predicción

```python
from sklearn.preprocessing import LabelEncoder
col = ['title', 'director', 'rating','listed_in','description', 'cast']
encoder = LabelEncoder()
df[col] = df[col].apply(encoder.fit_transform)

df.head()
```

# Seleccionamos variables independientes y dependientes

```python
X = df.iloc[:,[1,2,3,9,10]]
Y = df.iloc[:,7]
```

```python
X
```

```python
Y
```

# Datos de entrenamiento y prueba

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.25, random_state=0)
```

# Generar modelo

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(min_samples_leaf=2)
classifier.fit(X_train, y_train)
```

```python
classifier.score(X_train,y_train)
```

```python
y_pred = classifier.predict(X_test)
y_pred
```
