# Grafica

Graficamos un scatter plot para observar en general la cantidad de películas que tiene cierta clasificación con respecto a su año de salida identificandolos con su categoría.

```python
sns.set_style(style='whitegrid')
sns.set(rc={'figure.figsize':(14,10)})
sns.scatterplot(data=df, x=df['release_year'], y=df['rating'], hue=df['listed_in'])

plt.title('Categoría de las peliculas/series con respecto a clasificación y año de salida')
plt.xlabel('Año de salida')
plt.ylabel('Clasificación')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0)

plt.show()
```

# Preprocesamiento de datos

Esto se hace debido a que en el dataset hay campos que se encuentran nulos, para que no haya problemas al realizar las predicciones se llena los datos nulos con NaN especificando que no se agregó un dato en esa campo.

```python
df.fillna('NaN')
```

Se utiliza label encoder para intercambiar los valores de cadena a numérico en las columnas que se utilizarán para aplicar el modelo y realizar la predicción

```python
from sklearn.preprocessing import LabelEncoder
col = ['director', 'rating','listed_in','description', 'cast']
encoder = LabelEncoder()
df[col] = df[col].apply(encoder.fit_transform)

df.head()
```

# Seleccionamos variables independientes y dependientes

Las variables independientes que utilizaremos es la columna de director, cast, listed_in y descripción; como variable independiente se utilizará el rating para predecir la clasificación que se asignará.

```python
X = df.iloc[:,[3,4,9,10]]
Y = df.iloc[:,7]
```

```python
X
```

```python
Y
```

# Datos de entrenamiento y prueba

Conseguimos los valores de entrenamiento y prueba, asignando las variables independientes y varaible dependiente con un 25% de tamaño de prueba y random_state de 0.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.25, random_state=0)
```

# Generar modelo

Utilizamos el algoritmo de árboles de decisión para generar el modelo con los parámetros que están por defecto solo plasmando el mínimo de muestras en la hoja que deben de ser 2 muestras como mínimo para estar en un nodo hoja

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(min_samples_leaf=2)
classifier.fit(X_train, y_train)
```

Obtenemos el score del modelo

```python
classifier.score(X_train,y_train)
```

# Predicción


Se utiliza el método predict a partir de los valores prueba para obtener la predicción del rating

```python
y_pred = classifier.predict(X_test)
y_pred
```
